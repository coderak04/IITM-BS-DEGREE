```python
import pandas as pd
data=pd.read_csv('Model_Building_1.csv')
data.head()
```

```python
features=data.drop(columns=['Sale Price'])
target=data['Sale Price']
x_train,x_test,y_train,y_test=features[0:8272],features[8272:],target[0:8272],target[8272:]
```

```python
from sklearn.linear_model import LinearRegression
```

Apply LinearRegression on the train dataset(X_train and y_train). What is the R2 score on the test dataset(X_test and y_test). ( Upto 4 digits after decimal points)

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```


Enter the maximum value of 𝑅2 score you got using the LinearRegression model when computed with 5 folds from the training dataset (X_train and y_train) using cross_val_score. ( Upto 4 digits after decimal points) obtained.

```python
from sklearn.model_selection import cross_val_score
```

```python
cvs=cross_val_score(lr,x_train,y_train,cv=5)
round(cvs.max(),4)
```

Train Ridge and Lasso with random_state=27 and keep other parameters as default using train dataset. Which one has the least "mean squared error" for the test dataset.

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_squared_error
```

```python
rid=Ridge(random_state=27)
las=Lasso(random_state=27)
rid.fit(x_train,y_train)
las.fit(x_train,y_train)
y_pred1=rid.predict(x_test)
y_pred2=las.predict(x_test)
print(mean_squared_error(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred2))
```

create a pipeline of the PolynomialFeatures(interaction_only=True) as transformer and Lasso as an estimator.

Use GridSearchCV for tuning the hyperparameters of the created pipeline on training dataset.

Keep polynomial degree as : [1,2]
lasso alpha value to be taken as : [10,100,1000,10000]
scoring : neg_mean_absolute_error.
cv = 5
n_jobs = -1 (negative one) [it helps in using all the computational power to run this job]
(Note: Kindly ignore the warning.)

Mark the best alpha value you got using above instructions.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
lasso=Lasso()
pf=PolynomialFeatures(interaction_only=True)
p={'degree':[1,2]}
gs=GridSearchCV(pf,param_grid=p,scoring='neg_mean_absolute_error',cv=5,n_jobs=-1)

param={'alpha':[10,100,1000,10000]}
gscv=GridSearchCV(lasso,param_grid=param,n_jobs=-1,cv=5,scoring='neg_mean_absolute_error')
pipe=Pipeline(steps=[('pf',gs),('gscv',gscv)])
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe[-1].best_params_
```

Enter the best polynomial degree value you got using above instructions.

```python
pipe[0].best_params_
```

Construct a pipeline to apply following steps

STEP-1 Selecting 10 best features based on the f_regression score function.
STEP-2 Apply Lasso algorithm with random_state=27 on that

What is the max P-VALUE you got for any feature of the train dataset

```python
from sklearn.feature_selection import SelectKBest,f_regression
```

```python
kbest=SelectKBest(f_regression,k=10)
las1=Lasso(random_state=27)
pipe1=Pipeline(steps=[('kbest',kbest),('lasso',las1)])
pipe1.fit(x_train,y_train)
```

```python
max(pipe1[0].pvalues_)
```


What is the max F_STATISTICS score you got for any feature of the train dataset?
