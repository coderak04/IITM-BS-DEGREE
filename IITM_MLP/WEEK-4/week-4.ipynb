```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
X,y=fetch_california_housing(as_frame=True,return_X_y=True)
features=pd.DataFrame(X)
target=pd.DataFrame(y)
features.head()
```

```python
features.isna().sum()
```

For Que 3 to 6, split the california housing dataset into train and test sets.

[Hint: use appropriate API with random_state=0, shuffle=False and test_size=0.2]

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,random_state=0,shuffle=False,test_size=0.2)
```

If we use StandardScaler for preprocessing and LinearRegression for fitting the model with training set obtained in previous questions, what are the coefficients obtained in the predicted model?

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
lr=LinearRegression()
ss=StandardScaler()
pipe=Pipeline(steps=[('scale',ss),('linear_reg',lr)])
pipe.fit(x_train,y_train)
```

```python
coeff=pipe.named_steps['linear_reg'].coef_
intercept=pipe.named_steps['linear_reg'].intercept_
print('Coefficient= ',coeff)
print('Bias=',intercept)
```

What is the R squared value for predictions obtained using test set features?

```python
pipe.score(x_test,y_test)
```

If we use StandardScaler for preprocessing and LinearRegression for fitting the model, what is the root mean squared error value for predictions obtained using test set features?

```python
from sklearn.metrics import root_mean_squared_error,explained_variance_score,max_error,mean_absolute_error,mean_squared_error
```

```python
y_pred=pipe.predict(x_test)
```

```python
root_mean_squared_error(y_test,y_pred)
```

```python
explained_variance_score(y_test,y_pred)
```

```python
max_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
