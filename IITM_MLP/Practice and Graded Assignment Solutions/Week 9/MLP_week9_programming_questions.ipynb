(NAT)  **Answer: 6**

**Solution**

```python
import numpy as np
import sklearn
import pandas as pd

from sklearn.datasets import load_wine
df = load_wine()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.3, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state = 1)
model.fit(X_train, y_train)
print('Score on Training set is:',model.score(X_train,y_train))
print('Score on Testing set is:', model.score(X_test,y_test))
print('Tree Depth Level is :', model.get_depth())
```

#Practice Question 4: Enter the value of the 'score' on training set. 

(NAT) **Answer: 0.8145 (Range : 0.8135-0.8155)**

#Practice Question 5: Enter the value of the 'score' on testing set.

```python
import numpy as np
import sklearn
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import tree


from sklearn.datasets import load_wine
df = load_wine()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.3, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy', splitter = 'random', max_depth = 3, min_samples_split = 4, min_impurity_decrease = 0.1, random_state = 1)
model.fit(X_train, y_train)
print('Score on Training set is:',model.score(X_train,y_train))
print('Score on Testing set is:', model.score(X_test,y_test))

#plt the figure, setting a black background
plt.figure(figsize=(20,8), facecolor ='w')

feature_names = df.feature_names
#create the tree plot
a = tree.plot_tree(model,
                   #use the feature names stored
                   feature_names = feature_names,
                   #use the class names stored
                   
                   rounded = True,
                   filled = True,
                   fontsize=12)
#show the plot
plt.show()
```

```python
import numpy as np
import sklearn
import pandas as pd

from sklearn.datasets import load_wine
df = load_wine()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.3, random_state = 1)
```

#Graded Question 10: What is the value of the best max_features criteria after performing GridSearchCV?

(a) 'auto'

(b) 'sqrt'

(c) 'log2'

(MCQ)  **Answer: (a) auto**

```python
import numpy as np
import sklearn
import pandas as pd

from sklearn.datasets import load_diabetes
df = load_diabetes()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.3, random_state = 1)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor


estimator = DecisionTreeRegressor(max_depth = 4, random_state = 1)

estimator_model = GridSearchCV(estimator,
       {'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
        'splitter':['random','best'],
        'max_features': [4,5,6,'sqrt', 'log2'],
        'ccp_alpha': [0.001,0.01,0.02,0.05]},
        cv = 4, return_train_score = True)
estimator_model.fit(X_train, y_train)
score = estimator_model.score(X_test,y_test)
print(score)
estimator_model.best_params_
```
