```python
import pandas as pd
import sqlalchemy as sa

# Create SQLAlchemy engine to connect to MySQL database
engine = sa.create_engine("mysql+pymysql://guest:relational@db.relational-data.org/restbase")

# Function to execute SQL queries and return results as DataFrame
def query(sql):
    return pd.read_sql(sql, engine)

# Query to calculate percentage of bakery food_type in each city
sql_query = """
SELECT
    g.city,
    COUNT(*) AS total_restaurants,
    SUM(CASE WHEN g.food_type = 'bakery' THEN 1 ELSE 0 END) AS bakery_restaurants,
    ROUND((SUM(CASE WHEN g.food_type = 'bakery' THEN 1 ELSE 0 END) / COUNT(*)) * 100, 2) AS bakery_percentage
FROM
    generalinfo g
JOIN
    location l ON g.id_restaurant = l.id_restaurant
GROUP BY
    g.city;
"""

# Execute the query and store results in a DataFrame
results_df = query(sql_query)

# Display the results to understand the structure and values
print(results_df.head())

# Find the percentage of bakery food_type in Daly City
daly_city_percentage = results_df[results_df['city'] == 'daly city']['bakery_percentage'].values[0]

# Count cities with a higher percentage of bakery food_type than Daly City
cities_higher_percentage = results_df[results_df['bakery_percentage'] > daly_city_percentage]['city'].nunique()

# Print the number of cities with a higher percentage of bakery food_type than Daly City
print(f"Number of cities with a higher percentage of bakery food_type than Daly City: {cities_higher_percentage}")
```

```python
import pandas as pd
import sqlalchemy as sa

# Create SQLAlchemy engine to connect to MySQL database
engine = sa.create_engine("mysql+pymysql://guest:relational@db.relational-data.org/restbase")

# Function to execute SQL queries and return results as DataFrame
def query(sql):
    return pd.read_sql(sql, engine)

# Query to calculate percentage of cafe food_type in each city
sql_query = """
SELECT
    g.city,
    COUNT(*) AS total_restaurants,
    SUM(CASE WHEN g.food_type = 'cafe' THEN 1 ELSE 0 END) AS cafe_restaurants,
    ROUND((SUM(CASE WHEN g.food_type = 'cafe' THEN 1 ELSE 0 END) / COUNT(*)) * 100, 2) AS cafe_percentage
FROM
    generalinfo g
JOIN
    location l ON g.id_restaurant = l.id_restaurant
GROUP BY
    g.city;
"""

# Execute the query and store results in a DataFrame
results_df = query(sql_query)

# Display the results to understand the structure and values
print(results_df.head())

# Find the percentage of cafe food_type in Sunnyvale
sunnyvale_percentage = results_df[results_df['city'] == 'sunnyvale']['cafe_percentage'].values[0]

# Count cities with a higher percentage of cafe food_type than Sunnyvale
cities_higher_percentage = results_df[results_df['cafe_percentage'] > sunnyvale_percentage]['city'].nunique()

# Print the number of cities with a higher percentage of cafe food_type than Sunnyvale
print(f"Number of cities with a higher percentage of cafe food_type than Sunnyvale: {cities_higher_percentage}")
```
