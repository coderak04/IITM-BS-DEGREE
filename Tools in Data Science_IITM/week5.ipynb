```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


# Function to compute cosine similarity
def compute_cosine_similarity(embedding1, embedding2):
    return cosine_similarity([embedding1], [embedding2])[0][0]

# Get the embedding for "Beautiful"
embedding_beautiful = embeddings_dict['Beautiful']

# Lists of words
list1 = ['Creative', 'Agreement', 'Strategy', 'Happiness']
list2 = ['Invention', 'Beautiful', 'Champion', 'Consider']
list3 = ['Universe', 'Majority', 'Ambitious', 'Fantastic']
list4 = ['Medicine', 'Education', 'Different', 'Astronomy']

# Function to compute average cosine similarity for a list of words
def average_cosine_similarity(embedding_base, word_list):
    total_similarity = 0.0
    for word in word_list:
        embedding_word = embeddings_dict[word]
        total_similarity += compute_cosine_similarity(embedding_base, embedding_word)
    return total_similarity / len(word_list)

# Compute average cosine similarity for each list
avg_similarity_list1 = average_cosine_similarity(embedding_beautiful, list1)
avg_similarity_list2 = average_cosine_similarity(embedding_beautiful, list2)
avg_similarity_list3 = average_cosine_similarity(embedding_beautiful, list3)
avg_similarity_list4 = average_cosine_similarity(embedding_beautiful, list4)

# Print results
print(f"Average cosine similarity with 'Beautiful' for list 1: {avg_similarity_list1}")
print(f"Average cosine similarity with 'Beautiful' for list 2: {avg_similarity_list2}")
print(f"Average cosine similarity with 'Beautiful' for list 3: {avg_similarity_list3}")
print(f"Average cosine similarity with 'Beautiful' for list 4: {avg_similarity_list4}")

# Determine the list with the highest average cosine similarity
average_similarities = {
    'list1': avg_similarity_list1,
    'list2': avg_similarity_list2,
    'list3': avg_similarity_list3,
    'list4': avg_similarity_list4
}

max_list = max(average_similarities, key=average_similarities.get)
print(f"\nThe list with the highest average cosine similarity with 'Beautiful' is: {max_list}")
```
