```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
df = pd.read_parquet("flights.parquet")

# Filter the dataset
filtered_df = df[(df['DISTANCE'] < 300) & (df['SCHEDULED_DEPARTURE'] >= 13) & (df['SCHEDULED_DEPARTURE'] < 21)]

# Remove rows with NaN or inf values
filtered_df = filtered_df.dropna(subset=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'])

# Calculate the Pearson correlation coefficient
correlation, _ = stats.pearsonr(filtered_df['DEPARTURE_DELAY'], filtered_df['ARRIVAL_DELAY'])
print(correlation)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_parquet("flights.parquet")

# Filter for flights traveling less than 2400 miles with scheduled departure between 0:00 and 1:00
filtered_df = df[(df['DISTANCE'] < 2400) & (df['SCHEDULED_DEPARTURE'] >= 0) & (df['SCHEDULED_DEPARTURE'] < 1)]

# Calculate Q1 and Q3 for arrival delays
Q1 = filtered_df['ARRIVAL_DELAY'].quantile(0.25)
Q3 = filtered_df['ARRIVAL_DELAY'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Count the number of outlier arrival delays
outliers = filtered_df[(filtered_df['ARRIVAL_DELAY'] < lower_bound) | (filtered_df['ARRIVAL_DELAY'] > upper_bound)]
outlier_count = outliers.shape[0]

print(outlier_count)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression


df = pd.read_csv("house-rent.csv")

# Filter for Unfurnished houses in Kolkata and Chennai
kolkata_unfurnished = df[(df['City'] == 'Kolkata') & (df['Furnishing Status'] == 'Unfurnished')]
chennai_unfurnished = df[(df['City'] == 'Chennai') & (df['Furnishing Status'] == 'Unfurnished')]

# Function to perform linear regression and get the slope
def get_slope(df):
    X = df[['Size']].values  # Independent variable
    y = df['Rent'].values    # Dependent variable
    
    model = LinearRegression()
    model.fit(X, y)
    
    slope = model.coef_[0]
    return slope

# Get the slope for Kolkata
kolkata_slope = get_slope(kolkata_unfurnished)

# Get the slope for Chennai
chennai_slope = get_slope(chennai_unfurnished)

# Calculate the absolute difference
slope_difference = abs(kolkata_slope - chennai_slope)
print(slope_difference)
```
